```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets 
```

```python
x,y= datasets.make_blobs(n_samples=1500,n_features=2,
   random_state=10)
```

```python
from sklearn.cluster import MeanShift,estimate_bandwidth
```

```python
bandwith=estimate_bandwidth( x,
    quantile=0.2,
    n_samples=1000,
    random_state=0,
    n_jobs=None,)
```

```python
cluster=MeanShift(bandwidth=bandwith)
```

```python
cluster.fit(x)
```

```python
lable=cluster.labels_
```

```python
cluster.predict(x)
```
